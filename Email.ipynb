```python
dictNameEmail = {}

def displayMenu():
    #Choices to input for users
    print ('Look up email address (l)')
    print ('Add a new name and email (a)')
    print ('Update an existing email (u)')
    print ('Delete an existing email (d)')
    print ('Quit (q)')
    print ()

    to_do = str(input('Enter what you wish to do: '))
    while to_do not in "laudq":
        print ("Enter one of above options.")
        to_do = str(input('Enter what you wish to do: '))
        print ()
    return to_do


def lookupEmail(dictNameEmail):
    name = str(input('Enter the name: '))
    print (dictNameEmail.get(name, 'Not found.'))
    print()
    
#defining addEmail function
def addEmail(dictNameEmail):
    #ask the user for a name
    name = str(input('Enter name: '))
    if name not in dictNameEmail:
        email = input('Enter email: ')
        dictNameEmail[name] = email
        print()
    else:
        print ('The entry already exists.')
        print()
        
def updateEmail(dictNameEmail):
    name = str(input('Enter name: '))
    if name in dictNameEmail:
        email = input('Enter email: ')
        dictNameEmail[name] = email
        print()
    else:
        print ('Name  not found')
        print()
        
def deleteEmail(dictNameEmail):
    name = str(input('Enter name: '))
    if name in dictNameEmail:
        del dictNameEmail[name]
        print()
    else:
        print ('Name not found')
        print()

def main(): 
    choice = ''
    
    while choice != 'q':
        choice = displayMenu()

        if choice == 'l':
            lookupEmail(dictNameEmail)

        elif choice == 'a':
            addEmail(dictNameEmail)

        elif choice == 'u':
            updateEmail(dictNameEmail)

        elif choice == 'd':
            deleteEmail(dictNameEmail)       

if __name__ == '__main__':
    main()
```
